```python
# Important dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
import requests
from config import weather_api_key
```

```python
# Create a new set of 2,000 random latitudes and longitudes.

lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
for coordinate in coordinates:
    print(citipy.nearest_city(coordinate[0], coordinate[1]).city_name,
         citipy.nearest_city(coordinate[0], coordinate[1]).country_code)
```

```python
# Create a list for holding the cities.
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
# API Call
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"



# Build query URL
query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=" + units

# Show query URL (WARNING: shows API key)
#query_url
```

```python
print(query_url)
```

```python
city_weather = requests.get(query_url)
city_weather
```

```python
weather_response = requests.get(query_url)
weather_json = weather_response.json()

weather_json
```

```python
lat = weather_json['coord']['lat']
lng = weather_json['coord']['lon']
max_temp = weather_json['main']['temp_max']
percent_humidity = weather_json['main']['humidity']
percent_cloudiness = weather_json['clouds']['all']
wind_speed = weather_json['wind']['speed']
weather_description = weather_json['weather']

print(lat, lng, max_temp, percent_humidity, percent_cloudiness, wind_speed, weather_description)
```

```python

weather_data = []


print("Beginning Data Retrieval     ")
print("-----------------------------")


record_count = 1
set_count = 1


for i, city in enumerate(cities):
    
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    
#     city_url = url + "&q=" + city.replace(" ","+")
    city_url = f"http://api.openweathermap.org/data/2.5/weather?appid=1801d9e2aaf1aed11a57c4a862810681&q={city}&units=imperial"
    
    
    
    print(f"Processing Record {record_count} of Set {set_count} | {city}")

 
    record_count += 1
    
    
    try:
        
        city_weather = requests.get(city_url).json()
        
        city = city_weather['name']
        lat = city_weather['coord']['lat']
        lng = city_weather['coord']['lon']
        max_temp = city_weather['main']['temp_max']
        percent_humidity = city_weather['main']['humidity']
        percent_cloudiness = city_weather['clouds']['all']
        wind_speed = city_weather['wind']['speed']
        city_country = city_weather['sys']['country']
       


       
        weather_data.append({"City": city,
                             "Lat": lat,
                             "Lng": lng,
                             "Max Temp": max_temp,
                             "Humidity": percent_humidity,
                             "Cloudiness": percent_cloudiness,
                             "Wind Speed": wind_speed,
                             "Country": city_country})

        
        


    except:
        print("City not found. Skipping...")
        pass
    

print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")

```

```python
weather_data_df = pd.DataFrame(weather_data)
weather_data_df
```

```python
# Create the output file (CSV).
output_data_file = "weather_data/WeatherPy_Database.csv"
# Export the City_Data into a CSV.
weather_data_df.to_csv(output_data_file, index_label="City")
```
